```python
import pandas as pd
import numpy as np
```

```python
df_info = pd.read_json('contacts.json')
```

```python
df_info
```

```python
# 將Id轉為字串，方便後續設定變數
df_info['Id'] = df_info['Id'].astype(str)
```

```python
# 列出非空值列
df_Email = df_info[~df_info['Email'].isin([''])]
df_Phone = df_info[~df_info['Phone'].isin([''])]
df_OrderId = df_info[~df_info['OrderId'].isin([''])]
```

```python
# 以各欄位進行groupby，一樣名稱的 id 轉為list
df_Email_id = df_Email.groupby('Email')['Id'].apply(lambda x: list(x))
df_Phone_id = df_Phone.groupby('Phone')['Id'].apply(lambda x: list(x))
df_OrderId_id = df_OrderId.groupby('OrderId')['Id'].apply(lambda x: list(x))
```

```python
# 合併三個 table
df_merge1 = pd.merge(df_info,df_Email_id,how='left',on='Email',suffixes=['','_mail'])
df_merge2 = pd.merge(df_merge1,df_Phone_id,how='left',on='Phone',suffixes=['','_Phone'])
df_merge = pd.merge(df_merge2,df_OrderId_id,how='left',on='OrderId',suffixes=['','_OrderId'])
df_merge
```

```python
# 將NAN值轉為空list
df_merge['Id_mail'] = df_merge['Id_mail'].apply(lambda x: [] if x is np.nan else x)
df_merge['Id_Phone'] = df_merge['Id_Phone'].apply(lambda x: [] if x is np.nan else x)
df_merge['Id_OrderId'] = df_merge['Id_OrderId'].apply(lambda x: [] if x is np.nan else x)
df_merge
```

```python
# 將合併起來產生新欄位
df_merge['list']=df_merge['Id_mail']+df_merge['Id_Phone']+df_merge['Id_OrderId']
df_merge
```

```python
# 去重覆與重排列
df_merge['list'] = df_merge['list'].apply(lambda x : sorted(list(set(x))))
df_merge
```

```python
# 將兩個欄位轉為 dict 格式
area_dict = dict(zip(df_merge.Id, df_merge.Contacts))
area_dict
```

```python
# 將包含在List裡的值，取出dict裡的value並累加
def Contacts_fun(area_dict,list):
    total = 0
    for i in list:
        total = total + area_dict[i]
    return total
```

```python
df_merge['total']=df_merge.apply(lambda x : Contacts_fun(area_dict,x['list']), axis=1)
df_merge
```

```python
# 字串相加成想輸出的格式
df_merge['list_apply'] = df_merge.apply(lambda x : '-'.join(x['list']) + ',' + str(x['total']), axis=1)
df_merge
```

```python
#重新輸出
df_final = df_merge[['Id','list_apply']]
df_final.columns = ['ticket_id','ticket_trace/contact']
df_final
```
