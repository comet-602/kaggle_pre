### 分割資料

```python
# 將label為nan值的當成test資料
```

```python
test_data = df_new[df_new['label'].isnull()]
train_data = df_new[df_new['label'].notnull()]
```

```python
test_data.shape,train_data.shape
```

```python
x_train = train_data.iloc[:,:-1]
x_test = test_data.iloc[:,:-1]
y_train = train_data.iloc[:,-1]
y_test = test_data.iloc[:,-1]
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# 將label不為nan值的拆分
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_ver,y_train,y_ver = train_test_split(x_train,y_train,test_size = 0.2 , random_state=42)
```

```python
x_train.shape,x_ver.shape,x_test.shape,y_train.shape,y_ver.shape,y_test.shape
```

### 決策樹

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import time
```

```python
tStart = time.time()
tree = DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=10, random_state=10)
tree_clf = tree.fit(x_train, y_train)
score = tree_clf.score(x_ver,y_ver)
tEnd = time.time()
print('決策樹分數為{:.5f}:'.format(score))
print('計算時間:{:.2f}秒'.format(tEnd-tStart))
```

```python
from sklearn.metrics import roc_auc_score 

tree_clf_scores = tree_clf.predict_proba(x_test)[:,1]
df_submission['label'] = tree_clf_scores
df_submission
```
